```python
import pandas as pd 
import numpy as np
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
import matplotlib as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('Decisiontab2.csv')
```

```python
data
```

```python
data.columns
```

```python
lb=LabelEncoder()
```

```python
data['Weather']=lb.fit_transform(data[['Weather']])
```

```python
data['Temp']=lb.fit_transform(data[['Temp']])
```

```python
data['Humid']=lb.fit_transform(data[['Humid']])
```

```python
data['Wind']=lb.fit_transform(data[['Wind']])
```

```python
data['Play']=lb.fit_transform(data[['Play']])
```

```python
data
```

```python
x = data[["Weather","Temp","Humid","Wind"]]
```

```python
x
```

```python
y=data["Play"]
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=4)
```

```python
ad=AdaBoostClassifier(n_estimators=5,learning_rate=1)
ad
```

```python
ad.fit(xtrain,ytrain)
```

```python
ad.predict([[1,2,0,0]])
```

```python
ad.score(xtrain,ytrain)
```

```python
ypred=ad.predict(xtest)
```

```python
print("accuracy score=",accuracy_score(ypred,ytest))
```
