```python
import pandas as pd 
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import matplotlib as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv("Decisiontab.csv")
```

```python
data
```

```python
lb=LabelEncoder()
```

```python
data['Weather']=lb.fit_transform(data['Weather'])
data['Wind']=lb.fit_transform(data['Wind'])
data['Temp']=lb.fit_transform(data['Temp'])
data['Humid']=lb.fit_transform(data['Humid'])
data['Play']=lb.fit_transform(data['Play'])
```

```python
data
```

```python
x=data[['Weather','Temp']]
y=data['Play']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
kn=KNeighborsClassifier(n_neighbors=2)
```

```python
kn
```

```python
kn.fit(xtrain,ytrain)
```

```python
kn.predict([[1,0]])
```

```python
kn.score(xtest,ytest)
```

```python
plot_decision_regions(xtrain.to_numpy(),ytrain.to_numpy(),clf=kn)
```
