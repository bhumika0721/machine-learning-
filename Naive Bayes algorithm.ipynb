```python
import pandas as pd 
import numpy as np 
from sklearn.naive_bayes import GaussianNB , MultinomialNB , BernoulliNB
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split 
```

```python
data = pd.read_csv('COVIDNAVEBAYES.csv')
```

```python
data
```

```python
x= data[['Covid','Flu']]
```

```python
y= data[['Fever']]
```

```python
lb = LabelEncoder()
```

```python
x['Covid']= lb.fit_transform(x[['Covid']])
```

```python
x['Flu'] = lb.fit_transform(x[['Flu']])
```

```python
x
```

```python
y['Fever'] = lb.fit_transform(y[['Fever']])
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size =0.3, random_state = 3)
```

```python
bl = BernoulliNB()
```

```python
bl.fit(xtrain,ytrain)
```

```python
bl.predict([[0,1]])
```

```python
bl.predict_proba([[0,1]])
```
