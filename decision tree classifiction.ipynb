```python
import pandas as pd 
import numpy as np 
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split 
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data = pd.read_csv('Decisiontab.csv')
```

```python
data
```

```python
lb = LabelEncoder()
```

```python
data['Weather']=lb.fit_transform(data['Weather'])
data["Weather"]
```

```python
data['Temp']=lb.fit_transform(data['Temp'])
data["Temp"]
```

```python
data['Humid']=lb.fit_transform(data['Humid'])
data["Humid"]
```

```python
data['Wind']=lb.fit_transform(data['Wind'])
data["Wind"]
```

```python
data['Play']=lb.fit_transform(data['Play'])
data["Play"]
```

```python
x= data[['Weather','Temp','Humid','Wind']]
```

```python
y = data['Play']
```

```python
x
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=0.2 , random_state=5)
```

```python
dc = DecisionTreeClassifier(criterion="entropy")
```

```python
dc.fit(xtrain,ytrain)
```

```python
data
```

```python
a = dc.predict([[0,1,1,1]])
if(a==0):
    print('u cant play')
else:
    print('u can play')
```

```python
dc.score(xtest,ytest)
```

```python
plt.figure(figsize=(12,5))
tree.plot_tree(dc,feature_names=x.columns)
plt.show()
```
