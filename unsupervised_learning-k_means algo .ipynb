```python
import pandas as pd     
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import silhouette_score
```

```python
data=pd.read_csv("Kmeanssalry.csv")
```

```python
data
```

```python
scaler=MinMaxScaler()
scaler.fit(data[["Age"]])
data["Age"]=scaler.transform(data[["Age"]])
scaler.fit(data[["Income"]])
data["Income"]=scaler.transform(data[["Income"]])
```

```python
data
```

```python
km=KMeans(n_clusters=3)
km
```

```python
ypred=km.fit_predict(data[["Age","Income"]])           #number devided by cluster 
```

```python
data["Cluster"]=ypred
data
```

```python
dt1=data[data.Cluster==0]
dt2=data[data.Cluster==1]
dt3=data[data.Cluster==2]
plt.scatter(dt1["Age"],dt1["Income"],color="green")
plt.scatter(dt2["Age"],dt2["Income"],color="yellow")
plt.scatter(dt3["Age"],dt3["Income"],color="blue")
plt.xlabel("No of cluster")
plt.ylabel("CLuster score")
plt.legend()
```

```python
km.cluster_centers_
```

```python
dt1=data[data.Cluster==0]
dt2=data[data.Cluster==1]
dt3=data[data.Cluster==2]
plt.scatter(dt1["Age"],dt1["Income"],color="green")
plt.scatter(dt2["Age"],dt2["Income"],color="yellow")
plt.scatter(dt3["Age"],dt3["Income"],color="blue")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color="pink",marker="*")
plt.xlabel("No of cluster")
plt.ylabel("CLuster score")
plt.legend()
```
